```python
# dict - {key : value} --> not a sequence
  # key cant be duplicated, but a value can be.
  # key cant be a mutable datatype, value can be.
  # if a key is duplicated, it accept its but with the value doesn't remain the same,
  #it accept the latest values that is assigned to that key.

dct = {1:100, 'a':200,  (20,40): 'python', (20,40):'python', 1:"Swajal"}
print(dct)
print(dct['a'])  # call by key
print(dct[(20, 40)])
```

```python
# filter, map, reduce

# filter
# syntax - filter(func, seq)

lst = list(range(10,21))

# traditional code
out_lst =[]
for i in lst:
  if i%2 == 0:
    out_lst.append(i)
print(out_lst)

# Using filter
print(list(filter(lambda num: num%2==0, lst)))
```

```python
# map
# syntax - map(func, seq)

nlst = list(range(0,21,2))

# Traditional code
out_lst = []
for i in nlst:
  out_lst.append(int(i*1.5))
print(out_lst)

# Using map
print(list(map(lambda num: int(num*1.5), nlst)))
```

```python
# reduce
# Import before using
# from functools import reduce
# syntax - reduce(func, seq)

flst = list(range(1,200))

# Traditional code
out_num = 0
for i in flst:
  out_num += i
print(out_num)

# Using reduce
from functools import reduce
print(reduce(lambda num1, num2: num1+num2, flst))
```

```python
x1 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y1 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x1, y1, color='green', label='scatter point')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
x2 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y2 = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])


plt.scatter(x1, y1, color='hotpink', label='old data', marker='2', linewidth=8)
plt.scatter(x2, y2, color='green', label='new data', marker='s')

plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Scatter points comparision graph')
plt.legend(loc=1)

plt.grid(axis='x')
plt.show()
```

```python
csx = np.arange(1,11)
csy = np.random.randint(100,1000,(10,))
clr = np.array(['red','green','blue','yellow','violet','pink','gray','lavender','cyan','silver'])

plt.scatter(csx, csy, color=clr)
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Scatter points with user defined colors')

plt.colorbar()
plt.show()
```

```python
# Alpha is used to adjust the transperency of the dots in scatter plot

ssx = np.random.randint(100, size=(15))
ssy = np.random.randint(100, size=(15))
clr = np.random.randint(100, size=(15))
sze = np.random.randint(1000, size=(15))

plt.scatter(ssx, ssy, c=clr, s=sze, alpha=0.5,cmap='nipy_spectral')
plt.xlabel('X - Axis')
plt.ylabel('Y - Axis')
plt.title('Scatter points with different size and solor')

plt.colorbar()
plt.show()
```

```python
import numpy as np
import pandas as pd
import warnings
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt
```

```python
#importing data.
ted = pd.read_excel('/content/Financial_Sales.xlsx')
ted.head(5)
```

```python
# dimensions of the database
ted.shape[0]  # '0' is used when we need to know the total rows.
ted.shape[1]  # '1' is used when we need to know the total columns.
print(f"There are {ted.shape[0]} rows and {ted.shape[1]} columns.")
```

```python
# To find the data types and not null values
ted.info()
```

```python
num, obj = 0,0
for feature in ted.columns:
  if ted.dtypes[feature] != '0':
    num += 1
  else:
    obj += 1
print("Number of Numerical and Categories features:")
print(f'There are {num} numerical datatype and {obj} categories datatype')
```

```python
  #Percentage of isnull values

ted.isnull().mean()*100
```

```python
#Find number of duplicated values
print(f"There are {ted.duplicated().sum()} duplicate in the dataframe.")
```

```python
#When want to see duplicate values in a Tabular form.
ted[ted.duplicated()]
```

```python
# When want to remove columns from the dataset.

print(f'There are {ted.shape[1]} columns in the dataset.') #To see numbers of columns in the dataset.
feat = ['Month Name', "Month Number"] #Name of the columns which are to be remove.
for remove in feat:                   # Use the loop function to remove the columns from the dataset.
  del[remove]
print(f"There are {ted.shape[1]} columns in the dataset.")
```
